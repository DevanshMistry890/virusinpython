# Virus Program in python to create shortcuts

**Aim** : You have to create virus program that creates shortcuts of files that are in directory in which you place your program file.

> **Pyshortcuts** To create shortcut of file for any OS

[library](https://pypi.org/project/pyshortcuts/) [Github](https://github.com/newville/pyshortcuts)



---



```python
!pip install pyshortcuts 
from pyshortcuts import make_shortcut
```

```python
#python function to find & create shortcuts for file

import os #To listing directory

#search for target files in path
def search(path):
	filestoinfect = []
	filelist = os.listdir(path)
	for filename in filelist:
		
		#If it is a folder
		if os.path.isdir(path+"/"+filename):
			filestoinfect.extend(search(path+"/"+filename))
			
		#If it is a python script -> adding to list
		elif filename[-3:] == ".py":
			filestoinfect.append(path+"/"+filename)
   
	return filestoinfect

#Shortcuts to be made in the target files
def infect(filestoinfect):
  for i in range(len(filestoinfect)):
    make_shortcut(filestoinfect[i], name=filestoinfect[i], folder='/content/sample_data')
```
